```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify.csv')
df = df.drop(['Unnamed: 0'], axis = 1)
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
```

```python
top_artists = df.groupby('artist').count().sort_values(by = 'key', ascending = False)[:10].index.values
```

```python
top_10 = df[df['artist'].isin(top_artists)]
```

```python
dance = top_10.groupby('artist')['danceability'].mean().tolist()
energy = top_10.groupby('artist')['energy'].mean().tolist()
tempo = top_10.groupby('artist')['tempo'].mean().tolist()
valence = top_10.groupby('artist')['valence'].mean().tolist()
acousticness = top_10.groupby('artist')['acousticness'].mean().tolist()
```

```python
objects = top_artists
y_pos = np.arange(len(objects))
performance = dance

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.ylabel('Artist')
plt.xlabel('Danceability')
plt.title('Rates of danceability among top 10 frequest artists')

plt.show()
```
